```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target_names[iris.target]

```

```python
# Carregando a base de dados Iris
iris = load_iris()
X = iris.data

# Aplicando a PCA
pca = PCA()
principal_components = pca.fit_transform(X)

# Obtendo os pesos 
weights_df = pd.DataFrame(data=pca.components_.T, columns=[f'PC{i}' for i in range(1, X.shape[1]+1)], index=iris.feature_names)

# Exibindo os pesos
print(weights_df)
```

```python
# Carregando a base de dados Iris
iris = load_iris()
X = iris.data
target = iris.target

# Aplicando a PCA
pca = PCA()
principal_components = pca.fit_transform(X)

#DataFrame com as componentes principais
principal_df = pd.DataFrame(data=principal_components, columns=[f'PC{i}' for i in range(1, X.shape[1]+1)])
principal_df['target'] = target

#matriz de componentes principais
print(principal_df)

#variância explicada acumulativa
explained_variance_ratio_cumulative = pca.explained_variance_ratio_.cumsum()
print(explained_variance_ratio_cumulative)


plt.figure(figsize=(4, 3))  

plt.xlabel('Número de Componentes Principais')
plt.ylabel('Variância Explicada Acumulativa')
plt.title('Análise da Variância Explicada')

plt.tight_layout() 
plt.show()
```

```python
# Carregando a base de dados Iris
iris = load_iris()
X = iris.data
y = iris.target

# Aplicando a PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['target'] = y

#pairplot
sns.set(style="ticks")
g = sns.pairplot(pc_df, vars=['PC1', 'PC2'], hue='target', diag_kind="hist", height=2) 
plt.show()
```
